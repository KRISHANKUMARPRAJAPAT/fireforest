## Algerian Forest Fires Dataset
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

# 1.Data Cleaning

```python
# missing values
# we first check is there any missing values present in dataset.
dataset[dataset.isnull().any(axis = 1)]
```

The dataset is covered in two sets based on Region from 122th index, we can make a new column best on the region.

 **1 : "Bejaia Region Dataset"**
 
 **2 : "Sidi-Bel Abbes Region Dataset"**

```python
# Adding new column with region.
dataset.loc[:122,'Region'] = 0         # only bejaia region dataset present in this column.
dataset.loc[122:,'Region'] = 1         # only sidi-bel abbes region dataset present in this column.
df = dataset
```

```python
df.head()
```

```python
#checking the info()
df.info()
```

```python
# coverting the region datatypes into integer because it has only 0,1 values.
df['Region'] = df['Region'].astype(int)
```

```python
df.head()
```

```python
# agian checking the missing values
df.isnull().sum()
```

```python
# removing the null values from the df
df = df.dropna().reset_index(drop = True)
```

```python
# rechecking the null values
df.isnull().sum()
```

```python
# Calling specific rows data
df.iloc[[122]]             #this is the usefull info so remove this.
```

```python
# removing the 122nd row
df = df.drop(122).reset_index(drop = True)
```

```python
# now checking is real values available on the 122th row
df.iloc[[122]]                    # yes
```

```python
df.columns
```

```python
# fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

# Changing the requreed coulumns as integer data tyep

```python
df.columns
```

```python
df.head()
```

```python
# changing ['day', 'month', 'year', 'Temperature', 'RH', 'Ws'] == integer data type 
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
# checking the result
df.info()
```

```python
# chainging ['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI', 'Classes'] into float datatype 
# for this run a simple python code: 
obj = [feature for feature in df.columns if df[feature].dtypes == 'O']
```

```python
obj
```

```python
# changing the data type with a for loop funtion 
for i in obj:
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
# Now our data set in cleaned so we save it in to_csv format:
df.to_csv('Algerian_forest_fires_cleaned_dataset_UPDATE.csv',index = False)
```
